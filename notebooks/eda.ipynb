## Data Investigation

Investigating the shape of the dataset

```python
print(f"Users data has {users_df.shape[0]} rows and {users_df.shape[1]} columns")
print(f"Transactions data has {transactions_df.shape[0]} rows and {transactions_df.shape[1]} columns")
```

Investigating the data types of the columns in the dataset

```python
users_df.info()
```

```python
transactions_df.info()
```

Determining the cardinality (number of unique values) for each columns

```python
print(f"Number of rows in users data: {users_df.shape[0]}")
users_df.nunique().sort_values(ascending=False)
```

```python
print(f"Number of rows in transactions data: {transactions_df.shape[0]}")
transactions_df.nunique().sort_values(ascending=False)
```

Investigating the presence and number of null values

```python
users_df.isnull().sum().sort_values(ascending=False)
```

```python
transactions_df.isnull().sum().sort_values(ascending=False)
```

Determining the percentage of null values to the entire dataset

```python
for columns in users_df.columns:
    if users_df[[columns]].isnull().sum().values[0] > 0:
        print(f"{columns} has missing values and contributes to {round(users_df[[columns]].isnull().sum().values[0]/users_df.shape[0]*100, 2)}% of the total dataset")
```

```python
for columns in transactions_df.columns:
    if transactions_df[[columns]].isnull().sum().values[0] > 0:
        print(f"{columns} has missing values and contributes to {round(transactions_df[[columns]].isnull().sum().values[0]/transactions_df.shape[0]*100, 2)}% of the total dataset")
```

Investigating the presence and number of duplicates

```python
print('Number of duplicated rows in users data: %s'%users_df.duplicated().sum())
```

```python
print('Number of duplicated rows in transactions data: %s'%transactions_df.duplicated().sum())
```
